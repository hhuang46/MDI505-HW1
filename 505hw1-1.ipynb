```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
#define a function to get Cm and draw the plot
def get_Cm(x, y,nam):
    
    #calculate Cm
    regr = linear_model.LinearRegression()
    regr.fit(x.reshape(-1, 1), y)
    Cm, b = regr.coef_, regr.intercept_
    print(Cm,b)
    
    #draw plot
    plt.figure()
    plt.scatter(x,y)
    plt.plot(x, regr.predict(x.reshape(-1,1)), color='red', linewidth=1)
    plt.savefig("c_s-vs-1M_at"+nam)

    #draw log-log plot
    plt.figure()
    plt.scatter(x,y)
    plt.plot(x, regr.predict(x.reshape(-1,1)), color='red', linewidth=1)
    plt.xscale('log')
    plt.yscale('log')
    plt.savefig("logc_s-vs-log1M_at"+nam)
    return Cm,b
```

```python
# choose the data we need
data_file_name = "505hw1\specific-heat-and-atomic-mass-metals.txt"
data = np.loadtxt(data_file_name)
M_at = data[:, 0]
c_s = data[:, 1]
a = 1/M_at
```

```python
#calculate Cm and draw the plot
nam=""
get_Cm(a, c_s , nam)
```

Based on the Dulong–Petit rule, Cm = 3R=3*8.314=24.942≈25.006.
As a result, this value is expected.

```python
#add Be to data
M_at=np.append(M_at,9.012)
c_s=np.append(c_s,1.825)
a = 1/M_at
nam="-Be"
get_Cm(a, c_s,nam)
```
